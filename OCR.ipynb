```python
import cv2
import numpy as np
import pytesseract
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
pytesseract.pytesseract.tesseract_cmd = r"/usr/bin/tesseract"
```

## For clear image

```python
img1 = cv2.imread("image.jpg")
```

```python
text = pytesseract.image_to_string(img1)
```

```python
text
```

```python
plt.imshow(img1,cmap = plt.cm.gray)
```

## For not clear image

```python
img = cv2.imread("img2.jpg")
```

```python
img = cv2.resize(img,None,fx=0.5,fy=0.5)
```

```python
plt.imshow(img,cmap = plt.cm.gray)
```

```python
text = pytesseract.image_to_string(img)
```

```python
text
```

### Converting image to grey scale image

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
text = pytesseract.image_to_string(gray_img)
```

```python
text
```

Some imporovement is there.

```python
plt.imshow(gray_img,cmap = plt.cm.gray)
```

### Conveting image to black and white

```python
adaptive_threshold = cv2.adaptiveThreshold(gray_img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 85, 11)
```

```python
plt.imshow(adaptive_threshold,cmap = plt.cm.gray)
```
